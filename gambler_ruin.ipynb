```python
import numpy as np
import matplotlib.pyplot as plt


p=[0.65,0.58,0.53,0.5,0.47,0.42,0.35]

p= np.asarray(p)
fig = plt.figure()
for i in range(len(p)):
    q=1-p[i]
    print(q)
    N=20
    u_k=np.zeros((N+1))
    for k in range(N+1):
#         print(k)
        if p[i]==0.5: 
            u_k[k]=(N-k)/N
        else:
            u_k[k]=((q/p[i])**k-(q/p[i])**N)/(1-(q/p[i])**N)
    plt.plot(u_k)

    
    
    
    
plt.ylabel('$u_{k}$')
plt.xlabel('k')

plt.show()
```

```python
number_of_simulations = 50

# Create an array of wins of length equal to number_of_simulations
wins = [False] * number_of_simulations

# Fair Game
p= 0.5
N=50
k=10
ki=np.zeros((number_of_simulations+1))

i=0
while i < number_of_simulations and k >= 0 and k<N+1 :

    if p==0.5: 
        pn =(N-k)/N
    else:
        pn =((q/p[i])**k-(q/p[i])**N)/(1-(q/p[i])**N)
#     print([pn,k])
    p1_lose = np.random.uniform(0, 1) < pn
    ki[i]=k
    if p1_lose:
        k = k -1
    else:
         k = k + 1
            
    i=i+1
fig = plt.figure()   
plt.plot(ki[0:i])
```
