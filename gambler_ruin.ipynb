```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Probabilidad de ruina para una apuesta variable y diversas probabilidad de ganar una unidad
p=[0.65,0.58,0.53,0.5,0.47,0.42,0.35]

p= np.asarray(p)
fig = plt.figure()
for i in range(len(p)):
    q=1-p[i]
    #print(q)
    N=20
    u_k=np.zeros((N+1))
    for k in range(N+1):
#         print(k)
        if p[i]==0.5: 
            u_k[k]=(N-k)/N
        else:
            u_k[k]=((q/p[i])**k-(q/p[i])**N)/(1-(q/p[i])**N)
    plt.plot(u_k) 
plt.ylabel('$u_{k}$')
plt.xlabel('k')
plt.show()
```

```python
# Simulador de juego 
number_of_simulations = 50
wins = [False] * number_of_simulations

# Fair Game
p= 0.5
q=1-p
a=50
k=40
ki=np.zeros((number_of_simulations+1))
print(p)
i=0
while i < number_of_simulations and k >= 0 and k<a+1 :

    if p==0.5: 
        pn =(a-k)/a
    else:
        pn =((q/p)**k-(q/p)**a)/(1-(q/p)**a)
    
    p1_lose = np.random.uniform(0, 1) < pn
    ki[i]=k
    #print([pn,k,p1_lose])
    if p1_lose:
        k = k -1
    else:
         k = k + 1
            
    i=i+1
fig = plt.figure()   
plt.plot(ki[0:i])
```
