```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Probabilidad de ruina para una apuesta variable y diversas probabilidad de ganar una unidad
p=[0.65,0.58,0.53,0.5,0.47,0.42,0.35]

p= np.asarray(p)
fig = plt.figure(figsize=(8, 6))
for i in range(len(p)):
    q=1-p[i]
    #print(q)
    a=20
    dk=np.zeros((a+1))
    for k in range(a+1):
#         print(k)
        if p[i]==0.5: 
            dk[k]=k*(a-k)
        else:
            sk=(q/p[i])**k
            sa=(q/p[i])**a
            dk[k]= (1/(1-2*p[i])) * (k- (a*(1-sk))/(1-sa))  
            message = f"P= {p[i]}"
    plt.plot(dk,label=message) 

plt.ylabel('$d_{k}$')
plt.xlabel('k')
plt.legend(loc='upper right')
plt.show()
```
